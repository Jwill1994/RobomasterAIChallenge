```python
import Image
import numpy as np
```

```python
data = np.zeros((120,180), dtype=np.uint8)
```

```python
for i in range(9,110):
    for j in range(9,170):
        data[i,j] = 255
```

```python
def setblack(data,lists):
    for i in range(lists[1]-1,lists[3]): ### Y
        for j in range(lists[0]-1,lists[2]): ### X
            data[i,j] = 0
```

```python
#180,120
listss = [(34,30,54,35),(126,85,146,90),\
          (38,62,43,82),(137,38,142,58),\
          (75,90,80,110),(100,10,105,30),\
          (80,58,100,62)]
```

```python
for item in listss:
    setblack(data,item)
```

```python
image = Image.fromarray(data)
image.show()
```

```python
image.save("map.png")
```
